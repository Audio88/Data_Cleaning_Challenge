```python
#This cell of code is seperating PMID and PCMID from 'PMID/PMCID' and putting them into their own new columns

def find_pmcid(x):
    relist = re.findall('PMC[0-9]+', str(x))
    try:
        result = relist[0]
    except IndexError:
        result = None
    return result

def find_pmid(x):
    relist = re.findall('\d{8}', str(x))
    try:
        result = relist[0]
    except IndexError:
        result = None
    return result

df['PMCID'] = df['PMID/PMCID'].apply(find_pmcid)
df['PMID'] = df['PMID/PMCID'].apply(find_pmid)
```

```python
aggregations = {
    'Article title' : {
        'Articles Published': 'count'
    },
    'COST': {
        'cost_mean' : 'mean',
        'cost_median' : 'median',
        'cost_standard_deviation' : 'std'
    }
}
#removing 'FutureWarning'
warnings.simplefilter(action='ignore', category=FutureWarning)

newdf = df.groupby('Journal title').agg(aggregations)
newdf[[('COST', 'cost_mean'), ('COST', 'cost_median'),('COST', 'cost_standard_deviation')]] = newdf[[('COST', 'cost_mean'), ('COST', 'cost_median'),('COST', 'cost_standard_deviation')]].round(2)
newdf.sort_values(by=[('Article title', 'Articles Published')], ascending=False)
```
